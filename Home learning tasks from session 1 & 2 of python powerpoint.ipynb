Task 5 of HLT from python session 1 powerpoint, shown below:

```python


num1 = int(input("Please enter a number"))
num2 = int(input("Please enter another number"))

operator = input("Please enter anyone of the following operators: enter + for addition, enter - for subtraction, enter / for division, enter * for multiplication, or enter ** for taking to the power of.")

if operator=="+":
    cal = num1 + num2
    print(num1, "+", num2, "=", cal)
elif operator=="-":
    cal=num1-num2
    print(num1, "-", num2, "=", cal)
elif operator =="*":
    cal = num1*num2
    print(num1, "*", num2, "=", cal)
elif operator =="/":
    cal = num1/num2
    print(num1, "/", num2, "=", cal)
else:
    cal = num1**num2
    print(num1, "**", num2, "=", cal)
```

Doing the task in the 3rd bullet point of home learning task on page 25 of python session 2 powerpoint, shown below:

```python
#creating a procedure with passing parameters a and b
def procedure_2(a,b):
    num1 = a
    num2 = b

    operator = input("Please enter anyone of the following operators: enter + for addition, enter - for subtraction, enter / for division, enter * for multiplication, or enter ** for taking to the power of.")

    if operator=="+":
       cal = num1 + num2
       print(num1, "+", num2, "=", cal)
    elif operator=="-":
       cal=num1-num2
       print(num1, "-", num2, "=", cal)
    elif operator =="*":
       cal = num1*num2
       print(num1, "*", num2, "=", cal)
    elif operator =="/":
       cal = num1/num2
       print(num1, "/", num2, "=", cal)
    else:
       cal = num1**num2
       print(num1, "**", num2, "=", cal)
```
