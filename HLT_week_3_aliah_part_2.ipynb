Q6.Create two arrays a and b, stack these two arrays vertically use the  np.dot and np.sum to calculate totals, shown below

```python
a=np.array([4, 8, 12, 16, 20, 24, 28, 32, 36, 40, 44, 48, 52]) # 13 rows
b = np.array([5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65]) #13 rows

#Stacking array a vertically onto array b, shown below:
stack=np.vstack((a,b))

#creating a new array where each entities in the this array is the sum of a column in the stack array coressponding to its index value
sum_col = np.sum(stack, axis=0)
print("stack=", stack,"sum_col=", sum_col)

# Doing matrix multiplication of array a and b using the dot function, shown below

mult= np.dot(a,b)
print("mult=", mult)
```
